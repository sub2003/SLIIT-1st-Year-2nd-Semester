```python
def factorial(n):
    result=1
    if n==0:
        return result
    elif n>0:
        for i in range(1,n+1):
            result*=i
        return result
    else:
        return "Not defined"

print(factorial(5))

# OR

import math

print(math.factorial(5))
```

```python
# I.
population = 100
years = 10

pop_list = [population]

for year in range(1, years):
    population = population + 0.15 * population - 5
    pop_list.append(population)

for i, p in enumerate(pop_list, start=1):
    print("Year", i, ":", round(p,2))

# II.
population = 100
years = 10

total_increase = 0

for year in range(1, years+1):
    new_population = population + 0.15 * population - 5
    increase = new_population - population
    
    total_increase += increase
    population = new_population

print("Total population increase:", round(total_increase,2))

# III.
population = 100

for year in range(1, 50):
    if population > 250:
        print("Population first exceeds 250 in year:", year)
        break
    population = population + 0.15 * population - 5
```
