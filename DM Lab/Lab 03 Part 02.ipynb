```python
def f(x):
    if(x==0):
        raise ValueError("Domain Error: x cannot be 0 for f(x)")
    
    return 1/x
f(10)
    
```

```python
f(0)
```

```python
import math
def g(x):
    if(x<=0):
        raise ValueError("Domain Error: x cannot be less than 0 or equal to 0 for g(x)")
    return math.sqrt(x)

g(-1)
```

```python
g(0)
```

```python
g(1)
```

```python
g(3)
```

```python
def is_in_domain(func,x):
    try:
        func(x)
        return True
    except ValueError:
        return False
is_in_domain(f,100)
```

```python
is_in_domain(g,0)
```

```python
is_in_domain(g,5)
```

```python
g(8)
```
