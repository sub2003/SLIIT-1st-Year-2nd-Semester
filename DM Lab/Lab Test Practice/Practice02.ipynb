```python
DeviceIDs=[501,502,503,504]

def IPAddToDevices(IDs):
    assigned_ip={}
    base_ip="192.168.1"
    last_digit=2;
    
    for i in IDs:
        assigned_ip[i]=base_ip+str(last_digit)
        last_digit+=1
    return assigned_ip

addedIP=IPAddToDevices(DeviceIDs)

print(addedIP)

for device,ip in addedIP.items():
    print(f"{device}: {ip}")
        
```

```python
import math
def p(x):
    return x**3-4*x+1
def q(x):
    if(x<=0):
        raise ValueError("Domain Error : x cannot be less than 0 for q(x).")
    return 1/math.sqrt(x)
def r(x):
    if(x<=-1):
        raise ValueError("Domain Error : x cannot be less than -1 for r(x).")
    return math.log(x+1)
functionList=[p,q,r]

def evaluate_functions(func_list, x):
    result=list();

    for func in func_list:
        try:
            result.append(func(x))
        except ValueError:
            result.append("Domain Error")
    return result

print(evaluate_functions(functionList,0))
```

```python
'''
Python Variable Scope & Functions – Summary Points
Variables
A variable created inside a function is called a local variable.
A variable created outside a function is called a global variable.
Local variables work only inside the function.
Global variables work throughout the whole program.
Global Keyword
You can read a global variable inside a function without using global.
To change a global variable inside a function, you must use global.
global x means use the global variable x, not create a local one.
If you assign a value after using global, it changes the global variable.
A global variable always stores the last assigned value.

Example idea:

y = 10
y = 2
y = 10
Final y = 10
Nonlocal Keyword
nonlocal is used in nested functions.
It modifies the variable from the outer function, not global.
LEGB Rule (Very Important)

Python searches variables in this order:
12. L – Local
13. E – Enclosing
14. G – Global
15. B – Built-in

So Python searches:

Local → Enclosing → Global → Built-in
Functions
Functions are defined using def.
Functions can be stored in variables.
Functions can be stored in lists.
Functions can be passed as arguments to other functions.
Functions can return other functions.
Exceptions
Errors are handled using try and except.
You can create your own error using raise.

Example:

raise ValueError("Error message")
Most Important Rules (Remember These)
⭐
Variables inside functions are local by default.
Global variables can be read inside functions.
To modify global variable → use global.
Python uses LEGB rule to find variables.
Last assignment decides the variable value.
Functions can be treated like variables.
Errors handled using try–except.
'''
print()
```
