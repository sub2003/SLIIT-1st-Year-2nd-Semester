```python
def sum(n):
    count=1
    total=0
    while count<n:
        total=total+count
        count=count+1
    return total

print(sum(5))
```

```python
def sum_of_digits(n):
    if n==0:
        return 0
    else:
        return (n%10)+sum_of_digits(n//10)

print("sum_of_digits(1234) =", sum_of_digits(1234))
```

```python
def gcd(a,b):
    if b==0:
        return a
    else:
        return gcd(b,a%b)

print("gcd(48, 18) =", gcd(48, 18))
print("gcd(48, 20) =", gcd(50, 21))
```

```python
def sum_even(n):
    if n < 2:
        return 0
    elif n % 2 != 0:
        return sum_even(n - 1)
    else:
        return n + sum_even(n - 2)


print("sum_even(8) =", sum_even(8))
print("sum_even(11) =", sum_even(11))
print("sum_even(10) =", sum_even(10))
```

```python
def savings(n):
    if n == 1:
        return 1000
    elif n == 2:
        return 1050
    else:
        return 1.05 * savings(n - 1) + savings(n - 2)


print("savings(1) =", savings(1))
print("savings(2) =", savings(2))
print("savings(3) =", savings(3))
print("savings(4) =", savings(4))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)


def sum_factorials(n):
    if n == 1:
        return 1
    else:
        return factorial(n) + sum_factorials(n - 1)


print("sum_factorials(4) =", sum_factorials(4))
```
