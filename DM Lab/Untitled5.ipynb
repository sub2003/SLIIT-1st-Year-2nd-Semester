```python
A = {"U1", "U2", "U3", "U4", "U5", "U6"}
B = {"U3", "U4", "U5", "U7"}
C = {"U5", "U7"}

# I.
if(C.issubset(A)==True):
    print("C is a subset of A")
else:
    print("C is not a subset of A")
# II.
print("The users who have both email and file server access : ",A.intersection(B))
# III.
print("The users who have file server access but not admin rights : ",B.difference(C))
# IV.
if(A.isdisjoint(C)==True):
    print("A and C are disjoint sets")
else:
    print("A and C are not disjoint sets.")
# V.
print("The users who have at least one email or file server access : ",A.union(B))
# VI
print("The users who have either email or file server access, but not both : ",(A.union(B)).difference(A.intersection(B)))
```
