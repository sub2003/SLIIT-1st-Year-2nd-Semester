```python
String=("Aththanayaka","Boralage","Chandana","Subhanu","Ravisankha","Aththanayaka")  #Tuple

for i in String:
    print(i,end=" ")

String[1]=1#'tuple' object does not support item assignment
# tuple is immutable

for i in String:
    print(i,end=" ")
```

```python
Person={"Name","Subhanu","Ravisankha","Aththnayaka","12345"}

for i in Person:
    print(i,end=" ")

print()

Person1={"Name":"Subhanu","4th name":"Subhanu","5th name":"Ravisankha","6th name":"Aththnayaka","id":"12345"}
for k,v in Person1.items():
    print(v,end=" ")

print()

for s,r in Person1.items():
    print(k,":",v,end=" ")

print()

print("{",end="")
for s,r in Person1.items():
    
    print(s,":",r,end=" ")
    if(Person1[-1]!=Person1.items()):
        print(",")

print("}")
```
