```python
weight=float(input("Enter the weight(kg) : "))
height=float(input("Enter the height(m) : "))

BMI=weight/(height*height)

if(BMI<18):
    print("Underweight")
elif(BMI<=25):
    print("Normal")
else:
    print("Overweight")

```

```python
name=input("Student's name : ")
subject=[0,0,0]
total=0.0
for i in range(3):
    subject[i]=float(input(f"Enter subject {i+1} marks : "))
    total=total+subject[i]
percentage=total/len(subject)

grade=""
if(percentage>=75):
    grade='A'
elif(percentage>=60):
    grade='B'
elif(percentage>=45):
    grade='C'
elif(percentage>=30):
    grade='D'
else:
    grade='F'

print("----Mark Sheet----")
print("Student name : ",name)
for i in range(3):
    print(f"subject {i+1} marks = ",subject[i])
print("total marks = ",total)
print("Percentage = ",percentage)
print("Grade = ",grade)
```
