```python
scores = [78, 85, 95, 62, 88, 75, 95, 85, 62, 90]
# I.
uniqueScores=[]
for i in scores:
    if i not in uniqueScores:
        uniqueScores.append(i) 
        
        
print("Unique scores : ",uniqueScores)

# II.
min=scores[0]
for i in range(0,len(scores)):
    if(min>scores[i]):
        min=scores[i]
while min in scores:
    scores.remove(min)


scores.append(99)
print(scores)

# III.
highestScores=[]
for a in range(0,3):
    max=scores[0]
    #highestScores=[]
    for i in range(0,len(scores)):
        if(max<scores[i]):
            max=scores[i]
    highestScores.append(max)
    scores.remove(max)

print("The top 3 highest scores in the unique scores(in descending order) : ",highestScores)
```

```python
products = { 101: ["Laptop", "Electronics", 15, 55000],
             102: ["Table", "Furniture", 30, 4500],
             103: ["Pen", "Stationery", 100, 20],
             104: ["Mobile Phone", "Electronics", 25, 15000],
             105: ["Chair", "Furniture", 50, 2500],
             106: ["Notebook", "Stationery", 200, 50],
             107: ["Headphones", "Electronics", 40, 2500],
             108: ["Water Bottle", "Kitchenware", 60, 350] }

# I.
Electronics=[]
for i in products.values():
    if(i[1]=="Electronics"):
        Electronics.append(i[0])
print("The names of the product of type Electronics : ",Electronics)

# II.
Furnitures=[]
for i in products.values():
    if(i[1]=='Furniture' and i[3]>3000):
        Furnitures.append(i[0])
print("The names of those products which are of type 'Furniture' and cost more than 3000 : ",Furnitures)

# III.
productIds=[]

for key, i in products.items():
    if(i[2]<50):
        productIds.append(key)
    
print("The product ID of the product with quantity less than 50 : ",productIds)
```

```python
sale_event_a = {1001, 1002, 1003, 1004, 1005, 1006}
sale_event_b = {1004, 1005, 1006, 1007, 1008, 1009}

# I.
intersection=sale_event_a.intersection(sale_event_b)
print("The customers who participated in both sales : ", intersection)

# II.
print("The customers who participated only in Sale Event A : ",sale_event_a.difference(intersection))

# III.
print("The customers who participated only in Sale Event B : ",sale_event_a.difference(intersection))

# IV.
union=sale_event_a.union(sale_event_b)
print("The customers who participated in at least one sale : ",union)

# V.
print("The customers who participated in exactly one sale : ",union-intersection)
```

```python
fever = {201, 202, 203, 204, 205, 206, 207}
cough = {203, 204, 205, 208, 209, 210}
fatigue = {202, 204, 206, 209, 211, 212}

# I.
intersection= fever.intersection(cough,fatigue)
print(intersection)

# II.
difference=fever.difference(cough,fatigue)
print(difference)

# III.

fc=fever.intersection(cough)
cfa=cough.intersection(fatigue)
faf=fatigue.intersection(fever)

twoOfThreeSymptoms=fc|cfa|faf

print(len(twoOfThreeSymptoms))

# IV.
fluTesting=fc|faf
print(fluTesting)
```

```python
F = {"React", "Angular"}
B = {"Node.js", "Django"}

# I.
from itertools import product
stacks = list(product(F, B))

print("List of all possible ordered pairs : ",stacks)

# II.

from itertools import combinations

F = {"React", "Angular"}
B = {"Node.js", "Django"}

D = list(F | B)

print("All possible deployment configurations:")

for r in range(len(D) + 1):
    for s in combinations(D, r):
        print(s)
```
