```python
F = { "React ", "Angular" }
B = { "Node.js" , "Django" }


D={"Docker","Nginx"}

tech_stacks={(f,b)for f in F for b in B}  #crtesian product F&B
print(tech_stacks)

from itertools import product

tech_stacks=set(product(F,B))
print(tech_stacks)
```

```python
from itertools import combinations

power_set=[]

for r in range (len(D)+1):
    for combo in combinations(D,r):
        power_set.append(set(combo))

print(power_set)
```
