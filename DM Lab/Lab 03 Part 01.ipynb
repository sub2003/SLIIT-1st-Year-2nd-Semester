```python
def ip_address_assignment(address_List):
    ip_map = {}
    base_ip = "192.168.1."
    last_digit = 2

    for address in address_List:
        ip_map[address] = base_ip + str(last_digit)
        last_digit += 1

    return ip_map


D = [501,502,503,504]

ip_add = ip_address_assignment(D)
print(ip_add)
for device, ip in ip_add.items(): #
    print(f"{device}: {ip}")
    
```

```python
def grade(mark):
    if(mark>=75):
        grade='A'
    elif(mark>=60):
        grade='B'
    elif(mark>=40):
        grade='C'
    else:
        grade='F'

    return grade

for a in range(5):
    while(True):
        marks=int(input("Enter subject "+str(a+1)+" marks :"))
        if(marks>100 or marks<0):
            continue
            
        else:
            print(grade(marks))
            break
```

```python
def is_in_domain(func,x):
    try:
        func(x)
        return True
    except ValueError:
        return False
```

```python
def count_vowels(s):
    vowelCount=0
    #s=input("Enter a string : ")
    for i in s:
        vowels=['a','A','e','E','i','I','o','O','u','U']
        if i in vowels:
            vowelCount+=1
    return vowelCount
print(count_vowels("hello world"))
```

```python
import math
def p(x):
    return x**3-4*x+1
def q(x):
    if(x<=0):
        raise ValueError("Domain Error : x cannot be less than 0 for q(x).")
    return 1/math.sqrt(x)
def r(x):
    if(x<=-1):
        raise ValueError("Domain Error : x cannot be less than -1 for r(x).")
    return math.log(x+1)

def evaluate_function(func_list,x):
    results=[]
    for func in func_list: 
        try:
            results.append(func(x))
        except ValueError:
            results.append("Domain Error")
    return results

functions=[p,q,r]

x=0

print(evaluate_function(functions,x))
```
