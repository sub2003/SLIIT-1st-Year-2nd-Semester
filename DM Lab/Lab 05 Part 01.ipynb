```python
# Q2
import itertools

lett = ["a","b","c","d"]
nums = [0,1,2,3,4]

characters = lett + nums
passwords = list(itertools.product(characters, repeat=5))

newlist = []
for i in passwords:
    letter_count = sum(1 for j in i if j in lett)
    digit_count = sum(1 for j in i if j in nums)
    if letter_count == 2 and digit_count == 3:
        newlist.append(i)

print(newlist)
print("Unique password count", len(newlist))
```

```python
#Q1
import itertools

books = ["R","R","R","G","G","B","O"]
permu = set(itertools.permutations(books, 7))
print(len(permu))

valid = []
for p in permu:
    s = "".join(p)
    if "RRR" in s and "GG" in s:
        valid.append(s)

print(len(valid))
print(valid)
```

```python
#Q2
import itertools

num = [1,2,3,4,5]
numbers = list(itertools.permutations(num, 3))

newlist = []
for i in numbers:
    joined = "".join(map(str, i))
    newlist.append(joined)

count = 0
for j in newlist:
    val = int(j)
    if val > 500:
        count += 1

print("three-digit numbers can be created =", len(newlist))
print("three-digit numbers that greater than 500 =", count)
```

```python
#Q3
import itertools

moves = ["R","R","R","U","U","U"]
paths = set(itertools.permutations(moves))
print(paths)

count = 0
count_avoid = 0

for p in paths:
    x = 0
    y = 0
    through2x2 = False

    for move in p:
        if move == "R":
            x += 1
        else:
            y += 1

        if (x, y) == (2, 2):
            count += 1
            through2x2 = True

    if not through2x2:
        count_avoid += 1

print("Paths through (2,2):", count)
print("Paths that avoid (2,2):", count_avoid)
```

```python
#Q4
import itertools

computeservers = ['C1','C2','C3','C4','C5','C6','C7','C8']
storageservers = ['S1','S2','S3','S4','S5']
backupservers = ['B1','B2']

all_servers = computeservers + storageservers + backupservers

all_combinations = list(itertools.combinations(all_servers, 11))
print("All combination", len(all_combinations))

computechoices = list(itertools.combinations(computeservers, 6))
storagechoices = list(itertools.combinations(storageservers, 4))
backupchoices = list(itertools.combinations(backupservers, 1))

valid_selections = []
for c in computechoices:
    for s in storagechoices:
        for b in backupchoices:
            valid_selections.append(list(c + s + b))

print("how many different valid selections are possible:", len(valid_selections))
```
